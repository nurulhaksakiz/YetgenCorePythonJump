# Problem 1
Bir YetGenli yeni aldığı bahçesinin etrafına çit çevirmek istiyor. Bahçenin etrafını çit ile çevirmek için gerekli çit uzunluğunu ve bahçenin alanını hesaplayan fonksiyonu yazınız.

````python
def bahce_cit_ve_alan_hesapla(uzunluk, genislik):
    """
    Verilen uzunluk ve genişliğe göre bir bahçenin etrafını çit ile çevirmek için gerekli
    çit uzunluğunu ve bahçenin alanını hesaplar.

    Args:
    uzunluk (float): Bahçenin uzunluğu.
    genislik (float): Bahçenin genişliği.

    Returns:
    tuple: Bir tuple objesi olarak (cit_uzunlugu, alan), çit uzunluğu ve bahçe alanı değerlerini içerir.
    """
    cit_uzunlugu = 2 * (uzunluk + genislik) 
    alan = uzunluk * genislik 

    return (cit_uzunlugu, alan)

uzunluk = 10
genislik = 5

cit_uzunlugu, alan = bahce_cit_ve_alan_hesapla(uzunluk, genislik)

print(f"Çit uzunluğu: {cit_uzunlugu}")
print(f"Bahçe alanı: {alan}")
````

# Problem 2
Kendisine gönderilen 2 sayı arasındaki tüm asal sayıları bulan fonksiyonu yazınız.

````python
def asal_sayilari_bul(baslangic, bitis):
    """
    Verilen iki sayı arasındaki tüm asal sayıları bulur.

    Args:
    baslangic (int): Asal sayıların aranacağı aralığın başlangıç noktası.
    bitis (int): Asal sayıların aranacağı aralığın bitiş noktası.

    Returns:
    list: Asal sayıların listesi.
    """
    asal_sayilar = [] 
    for sayi in range(baslangic, bitis+1): 
        if sayi > 1: 
            for i in range(2, sayi): 
                if (sayi % i) == 0: 
                    break 
            else: 
                asal_sayilar.append(sayi) 

    return asal_sayilar 
baslangic = 1
bitis = 20

asal_sayilar = asal_sayilari_bul(baslangic, bitis)

print(f"{baslangic} ve {bitis} arasındaki asal sayılar: {asal_sayilar}")
````

# Problem 3
YetGen liderlerinin kurdukları girişim unicorn oluyor. YetGen liderleri iş arkadaşlarının alacağı ücretlerini hesaplamak için bir fonksiyon yazdılar. Bu fonksiyonu kullanarak, YetGen liderlerinin iş arkadaşlarına verecekleri ücretleri hesaplayınız.

Ücret hesaplanması çalışma saatine bağlıdır. Bunun için kullanıcıdan çalışma saati bilgisinn alın. Sonrasında yapılacak ödemenin hesaplanacağı computePay() adlı bir fonksiyon oluşturun. Bu fonksiyonun içinde işlemler yapın. Çalışma saati 40 ve altındaysa saat başına ücret 10 TL’dir. 40 saatin üstünde çalışıldıysa saat başına ücret 15 TL sayılmaktadır.

```
Test etmek için 30 saat değerini giriniz,  sonuç 300 TL çıkmalı, 
ikincil test olarak 50 değerini giriniz,  sonuç 750 TL çıkmalı.
```

````python
def computePay(saat):
    """
    Verilen saat sayısına göre iş arkadaşına ödenecek ücreti hesaplar.

    Args:
    saat (float): İş arkadaşının çalıştığı saat sayısı.

    Returns:
    float: İş arkadaşına ödenecek ücret.
    """
    if saat <= 40: 
        ucret = saat * 10 
    else: 
        ucret = 40 * 10 + (saat - 40) * 15 
        
    return ucret 
saat = 30

ucret = computePay(saat)

print(f"{saat} saat çalışma sonucunda ödenecek ücret: {ucret} TL")

saat = 50

ucret = computePay(saat)

print(f"{saat} saat çalışma sonucunda ödenecek ücret: {ucret} TL")

````

# Problem 4
YetGenliler zirveye gitmek için toplanıyor. Zirveye gitmek için kiraladıkları araçlar 4'er kişiliktir. Gelecek kişi sayısı için Airtable üzerinde form açılıyor. Gelecek kişi sayısını alıp dört kişilik araçlara bölüp kalan kişi sayısını döndüren bir fonksiyon yazınız.

````python
def kalanKisiSayisi(kisiSayisi):
    """
    Verilen kişi sayısını 4 kişilik araçlara bölüp kalan kişi sayısını hesaplar.

    Args:
    kisiSayisi (int): Zirveye katılacak kişi sayısı.

    Returns:
    int: 4 kişilik araçlara bölündükten sonra kalan kişi sayısı.
    """
    kalan = kisiSayisi % 4
    return kalan 
kisiSayisi = 23

kalan = kalanKisiSayisi(kisiSayisi)

print(f"{kisiSayisi} kişi için 4 kişilik araçlara bölündükten sonra kalan kişi sayısı: {kalan}")

````

# Problem 6

[("xxx",1),("xxx",2),("xxx",9),("xxx",3), ("xxx",1),("xxx",13),("xxx",26),("xxx",4)] listesi içerisinde verilen elemanların her birini tuple içindeki 2. Elemana göre sıralayınız. 



````python
lambda tup: tup[1]
liste = [("xxx",1),("xxx",2),("xxx",9),("xxx",3),("xxx",1),("xxx",13),("xxx",26),("xxx",4)]
siralama_kriteri = lambda tup: tup[1]
liste_sirali = sorted(liste, key=siralama_kriteri)
print(liste_sirali)
````
