# Problem 1

Şimdi beraber kullanıcılardan aldığımız verileri işleyerek, onlara hem çıktı verecek, hem de aldığımız ve hesapladığımız verileri ileride kullanabilmek adına kaydedeceğiz.


 **Bu bir veri giriş platformu olacak**


 * kullanıcıya kaç giriş yapacağını soracağız
* kullanıcıdan isim-soyisim, yaş, boy, kilo bilgilerini alarak VKİ hesaplayacağız
* ardından bunları sınıflara göre kaydedeceğiz bir dosyaya kaydedeceğiz


Sınıflar ise şu şekilde: 
* 18, 5 kg/m.'nin altında olanlar: Zayıf
* 18.5 – 24, 9 kg/m. arasında olanlar: Normal kilolu
* 25 – 29, 9 kg/m. arasında olanlar: Fazla kilolu
* 30 – 39, 9 kg/m. arasında olanlar: Obez
* 40 kg/m.'nin üzerinde olanlar: İleri derecede obez (morbid obez), olarak görülür.

```python

n = int(input("Kaç kişinin bilgilerini gireceksiniz? "))

kullanicilar = []

for i in range(n):
    print(f"\n{i+1}. kişinin bilgilerini girin:")

    ad_soyad = input("Adınız ve soyadınız: ")
    yas = int(input("Yaşınız: "))
    boy = float(input("Boyunuz (metre cinsinden): "))
    kilo = float(input("Kilonuz (kilogram cinsinden): "))

    vki = kilo / (boy ** 2)

    if vki < 18.5:
        sinif = "Zayıf"
    elif 18.5 <= vki <= 24.9:
        sinif = "Normal kilolu"
    elif 25 <= vki <= 29.9:
        sinif = "Fazla kilolu"
    elif 30 <= vki <= 39.9:
        sinif = "Obez"
    else:
        sinif = "İleri derecede obez"

    kullanici = {
        "ad_soyad": ad_soyad,
        "yas": yas,
        "boy": boy,
        "kilo": kilo,
        "vki": vki,
        "sinif": sinif
    }

    kullanicilar.append(kullanici)


with open("kullanicilar.txt", "a") as f:
    for kullanici in kullanicilar:
        f.write(f"{kullanici['ad_soyad']} {kullanici['yas']} {kullanici['boy']} {kullanici['kilo']} {kullanici['vki']} {kullanici['sinif']}\n")
```

# Problem 2
**YetBar - Smoothie Bank**
---
YetBar spor salonunun girişinde bulunan, insansız bir sağlıklı içecek otomatıdır. Kullanıcı çeşitli girdiler smoothie içeceğine koyacağı ürünleri belirleyebilir, fiyat hesaplaması yaptırabilir ya da sadece menüyü görüntüleyebilir.

**Projenin Amacı:**

*   Birbirini etkileyen ve bir kod içerisinde birden fazla defa başvurulacak fonksiyonların oluşturulup, "fonksiyonel" proglamlamak.
*   Büyük programları yazarken, yazılımcı olarak yapabileceğimiz hatalardan kaçınmak için kontrol yöntemleri geliştirmek.
*   Kullanıcı inputlarını ve yaptığımız hesapları, raporlamak/loglamak amacıyla kaydetmek.

**Programdan Beklenen Çıktılar**
* Program kullanıcıya temelde 2 farklı işlem yapma imkanı sağlar.
> 1. Menüyü görüntüleme
> 3. Satın alma
* Satın alınan her içecek tarih, fiyat bilgisi ile kaydedilmeli ve her zaman ulaşılabilir olmalıdır.




```python
import datetime

class Smoothie:
    def __init__(self, name, price):
        self.name = name
        self.price = price

class YetBar:
    def __init__(self):
        self.smoothies = [Smoothie("Yeşil Smoothie", 7.50), Smoothie("Meyveli Smoothie", 8.50), Smoothie("Proteinli Smoothie", 10.50)]
        self.transactions = []
    
    def show_menu(self):
        print("SMOOTHIE MENÜSÜ")
        for i, smoothie in enumerate(self.smoothies):
            print(f"{i+1}. {smoothie.name} - {smoothie.price} TL")
    
    def buy_smoothie(self, smoothie_index):
        if smoothie_index > len(self.smoothies) or smoothie_index < 1:
            print("Hatalı seçim yaptınız. Lütfen tekrar deneyiniz.")
            return
        
        smoothie = self.smoothies[smoothie_index - 1]
        self.transactions.append({"tarih": datetime.datetime.now(), "isim": smoothie.name, "fiyat": smoothie.price})
        print(f"{smoothie.name} satın alındı. Toplam ödeme: {smoothie.price} TL")
    
    def save_transactions(self, file_name):
        with open(file_name, "w") as f:
            for transaction in self.transactions:
                f.write(f"{transaction['tarih']}\t{transaction['isim']}\t{transaction['fiyat']}\n")

yetbar = YetBar()

while True:
    print("\nLütfen bir seçim yapınız:")
    print("1. Menüyü görüntüle")
    print("2. Smoothie satın al")
    print("3. Çıkış yap")

    choice = int(input())

    if choice == 1:
        yetbar.show_menu()
    elif choice == 2:
        yetbar.show_menu()
        smoothie_choice = int(input("Lütfen satın almak istediğiniz smoothie'nin numarasını giriniz: "))
        yetbar.buy_smoothie(smoothie_choice)
    elif choice == 3:
        file_name = input("Kaydetmek istediğiniz dosya adını giriniz: ")
        yetbar.save_transactions(file_name)
        break
    else:
        print("Hatalı seçim yaptınız. Lütfen tekrar deneyiniz.")
```
