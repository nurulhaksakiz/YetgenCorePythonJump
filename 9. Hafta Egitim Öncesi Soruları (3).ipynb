# Problem 1
Vehicle sınıfından miras alan bir Bus sınıfı oluşturun. Bus.seating_capacity()'nin kapasite bağımsız değişkenine varsayılan olarak 50 değerini verin.

```
Output:
> The seating capacity of a bus is 50 passengers
```


````python
class Vehicle:
    def __init__(self, color, max_speed):
        self.color = color
        self.max_speed = max_speed

class Bus(Vehicle):
    def seating_capacity(self, capacity=50):
        print("The seating capacity of a bus is {} passengers".format(capacity))

bus = Bus('yellow', 80)
bus.seating_capacity()
````

# Problem 2


School_bus'ın aynı zamanda Vehicle sınıfının bir örneği olup olmadığını belirleyiniz.

````python
class Vehicle:
    def __init__(self, make, model, year):
        self.make = make
        self.model = model
        self.year = year

class Bus(Vehicle):
    def seating_capacity(self, capacity=50):
        return f"The seating capacity of a bus is {capacity} passengers"

class School_bus(Bus):
    def student_capacity(self, capacity=30):
        return f"The student capacity of the school bus is {capacity} students"
    
    
print(isinstance(School_bus(), Vehicle))
````
